```python
def df_value_counts(df, discrete_threshold=50):
    '''
    Функция, которая применяет value_counts ко всем столбцам
    '''
    for col in df.columns:
        s = df[col]
        n_unique = s.nunique(dropna=False)
        n_missing = s.isna().sum()
        print(f"\n================ {col} ================")
        if s.dtype == 'object' or n_unique <= discrete_threshold:
            print('Дискретный признак')
            print(s.value_counts(dropna=False))
        else:
            print('Непрерывный признак')
```

```python
def corr_mat(df):
    '''
    Функция, которая строит матрицы с учетом категориальных признаков
    '''
    matrix = df.phik_matrix() 
    plt.figure(figsize=(30, 30))
    sns.heatmap(matrix, annot=True, cmap='coolwarm')
    plt.show()
```

```python
def hist_plot_pie_plot(data):
    """
    Функция, которая автоматически строит подходящие диаграммы для каждого столбца
    """
    # Числовые и категориальные столбцы (исключаем id если есть)
    numeric_cols = data.select_dtypes(include=['int', 'float']).columns
    if 'id' in numeric_cols:
        numeric_cols = numeric_cols.drop('id')
    
    categorical_cols = data.select_dtypes(include=['object']).columns
    
    # Дискретные признаки
    discrete_cols = ['employment_years', 'supervisor_evaluation']
    
    # Графики для числовых столбцов, исключая дискретные
    for col in numeric_cols:
        if col not in discrete_cols:
            plt.figure(figsize=(12, 5))
            
            # Гистограмма
            plt.subplot(1, 2, 1)
            data[col].hist(bins=200)
            plt.title(f'Гистограмма {col}')
            plt.xlabel('Значения')
            plt.ylabel('Частота')
            
            # Ящик с усами
            plt.subplot(1, 2, 2)
            data[col].plot.box()
            plt.title(f'Ящик с усами {col}')
            plt.ylabel('Значения')
            
            plt.tight_layout()
            plt.show()
    
    # Строим круговые диаграммы для категориальных столбцов
    for col in categorical_cols:
        counts_norm = data[col].value_counts(normalize=True)
        counts = data[col].value_counts()
        
        # Круговая диаграмма с таблицей долей
        fig, ax = plt.subplots(figsize=(10, 10))
        counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, ax=ax)
        ax.set_title(f'Круговая диаграмма {col}')
        ax.set_ylabel('')
        
        # Таблица с долями
        table_data = counts_norm.reset_index().values
        table_col_labels = [col, 'Доля (%)']
        ax.table(cellText=table_data, colLabels=table_col_labels, loc='bottom', bbox=[0, -0.5, 1, 0.3])
        
        plt.tight_layout()
        plt.show()
        
        # Отдельная таблица с абсолютными значениями
        fig, ax = plt.subplots(figsize=(6, len(counts)*0.5))
        ax.axis('off')
        tb = ax.table(cellText=counts.reset_index().values, colLabels=[col, 'Количество'], loc='center', cellLoc='center')
        tb.auto_set_font_size(False)
        tb.set_fontsize(12)
        tb.scale(1.2, 1.2)
        plt.title(f'Таблица количеств для {col}')
        plt.tight_layout()
        plt.show()
    
    # Countplot для дискретных признаков
    for col in discrete_cols:
        if col in data.columns:
            plt.figure(figsize=(10, 6))
            sns.countplot(data=data, x=col, palette='viridis')
            plt.title(f'Countplot для {col}')
            plt.xlabel(col)
            plt.ylabel('Количество')
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()
```

## Обучние модели

Разделение на тренировочные и тестовые выборки без y_test

```python
X_train = heart_train.sort_index().drop(columns=['Heart Attack Risk (Binary)'])
y_train = heart_train.sort_index()['Heart Attack Risk (Binary)']
X_test = heart_test.sort_index()
```

Из-за того, что в целевом признаке классы не в балансе использую SMOTE, который синтетически добавит значений проигрывающему в кол-ве классу и поможет модели более эффективно обучиться

```python
sampler = SMOTE(random_state=42, k_neighbors=6)
```

```python
X_train_resample, y_train_resample = sampler.fit_resample(X_train, y_train) 
```

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', CatBoostClassifier(auto_class_weights='Balanced', random_state=42, verbose=False))
])
```

Пробовал разные модели, выбрал лучшую, макбук на 8гб с м1 не съел это все за раз

```python
param_grid = [
    # {
    #     'model': [LGBMClassifier(class_weight='balanced', random_state=42)],
    #     'model__num_leaves': [15, 31, 63],
    #     'model__max_depth': [-1, 5, 10],
    #     'model__learning_rate': [0.01, 0.1],
    #     'model__n_estimators': [100, 300, 500]
    # },
    # {
    #     'model': [SVC(class_weight='balanced', random_state=42, probability=True)],
    #     'model__C': [0.1, 1, 5, 10],
    #     'model__kernel': ['linear', 'rbf', 'poly'],
    # },
    #     {
    #     'model': [RandomForestClassifier(class_weight='balanced', random_state=42)],
    #     'model__n_estimators': [100, 300, 500],
    #     'model__max_depth': [5, 10, None],
    #     'model__min_samples_split': [2, 5, 10]
    # },
    {
        'model': [CatBoostClassifier(auto_class_weights='Balanced', random_state=42, verbose=False)],
        'model__depth': [3, 7, 10],
        'model__learning_rate': [0.01, 0.05, 0.1],
        'model__iterations': [100, 300, 500]
    }
]
```

```python
grid_search = GridSearchCV(
    pipeline,
    param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs = -1,
    verbose=2
)

grid_search.fit(X_train_resample, y_train_resample)
print(f'Лучшая модель:{grid_search.best_estimator_}')
```

```python
print(f'roc_auc лучший результат модели: {grid_search.best_score_:.2f}')
```

```python
grid_search_recall = GridSearchCV(
    pipeline,
    param_grid,
    cv=5,
    scoring='recall',
    n_jobs=-1,
    verbose=2
)
grid_search_recall.fit(X_train_resample, y_train_resample)
print(f'Лучшая модель:{grid_search_recall.best_estimator_}')
```

```python
print(f'recall лучший результат модели: {grid_search_recall.best_score_:.2f}')
```

**Промежуточный вывод:** обучил две модели с разынми метриками: roc-auc и recall
- roc-auc, чтобы посмотреть насколько в среднем вероятность найти больного выше вероятности найти здорового (78%, т.е. почти в 8 случах из 10 вероятность найти больного если он реально болен), для студента - нормальный результат, но вот для медицины как мне кажется, это может быть недопустимо
- recall, в медицине нестолько важно дать здоровому человеку ложное предсказание того, что он болен, нежели чем сказать больному человеку, что он здоров, поэтому надо отловить и нейтрализовать ошибки второго рода (FN), наче говоря крайне важно не пропустить реально больных. (62%, то есть модель выявляет 6 из 10 больных, остальных четверых модель пропускает)

## Проверка (фиктивная)

```python
best_model = grid_search_recall.best_estimator_
best_model.fit(X_train_resample, y_train_resample)
```

```python
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, y_train, 
    test_size=0.2, 
    stratify=y_train, 
    random_state=42
)
```

```python
best_model.fit(X_tr, y_tr)
y_proba_val = best_model.predict_proba(X_val)[:, 1]
```

```python
precisions, recalls, thresholds = precision_recall_curve(y_val, y_proba_val)
f1_scores = (2 * precisions[:-1] * recalls[:-1]) / (precisions[:-1] + recalls[:-1] + 1e-9)
best_idx = f1_scores.argmax()
best_threshold = thresholds[best_idx]
```

```python
print(f'Лучший порог по F1: {best_threshold:.2f}')
```

```python
y_proba_test = best_model.predict_proba(X_test)[:, 1]
y_pred_test = (y_proba_test > best_threshold).astype(int)
```

```python
y_proba_val = best_model.predict_proba(X_val)[:, 1]
y_pred_val = (y_proba_val > best_threshold).astype(int)
recall_val = recall_score(y_val, y_pred_val)
print(f'Recall на валидации при пороге {best_threshold:.2f}: {recall_val:.3f}')
```

```python
y_pred_val = (y_proba_val >= best_threshold).astype(int)
```

```python
precision = precision_score(y_val, y_pred_val)
recall = recall_score(y_val, y_pred_val)
f1 = f1_score(y_val, y_pred_val)
```

```python
print(f'Precision: {precision:.3f}')
print(f'Recall: {recall:.3f}')
print(f'F1: {f1:.3f}')
```

**Промежуточный вывод:**
- Решил сделать проверку из предоставленного датафрейма: из представленных данных сделал тренировочную и валидационную выборки, нашел порог по порогу округлил значения до 0 или 1, что и является предсказанием
- По метрикам: F1 показывает баланс между Precision и Recall (54% - говорит о балансе между метриками), Precision - низкий, т.е. из всех которым пердсказали, что они больные, действительно больными оказались только 40%, Recall - после установления порога модель нашла реально больных 90%. В рамках медицины такое соотношение метрик примелемо, но не идеально.

(62%, то есть модель выявляет 6 из 10 больных, остальных четверых модель пропускает)

- По метрикам: F1 показывает баланс между Precision и Recall (54% - говорит о балансе между метриками), Precision - низкий, т.е. из всех которым пердсказали, что они больные, действительно больными оказались только 40%, Recall - после установления порога модель нашла реально больных 90%.
